**10)Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe, first name, last name, OBP, order descending by OBP)**

```python
#Select data for 1977
a10 = batting_data[batting_data['yearID']==1977].reset_index(drop=True)

#Aggregate based on playerID
a100 = a10.groupby(["playerID", "nameFirst", "nameLast"],as_index=False)[['AB','BB','HBP','SH','SF','H']].sum()

#Calculate the PA
a100['PA'] = a100['AB'] + a100['BB'] + a100['HBP'] + a100['SH'] + a100['SF']

#Selecting players with PA greater than or equal to 500
pa10 = a100[a100['PA']>=500].reset_index(drop=True)

#Calculating OBP
pa10['OBP'] = ( pa10['H'] + pa10['BB'] + pa10['HBP'])/(pa10['AB']+ pa10['BB'] + pa10['HBP'] +pa10['SF'])

#Sort the values
pa10.sort_values('OBP',inplace=True,ascending=False)
#display
display(pa10.head(5).reset_index(drop=True))
```

**13)	Who led MLB with the most hits the most times?  And how many times?  (Dataframe, Number of hits)**

```python
#Aggregate with Player and Year
a13 = batting_data.groupby(['playerID','yearID'],as_index=False)[['H']].sum().copy()
a133 = a13.groupby('yearID',as_index=False)[['H']].max().copy()
a13_merged = pd.merge(a133,a13,how='left',on=['yearID','H']).copy()
#Find the number of times each player scored the most hits in a year
a13_merged['Number_of_Times']=a13_merged.groupby('playerID',as_index=False)['yearID'].count()['yearID']
a13_merged.sort_values('Number_of_Times',inplace=True,ascending=False)
#Displaying the result
display(a13_merged.reset_index(drop=True).head(1))
```

**24)Who is the youngest player to hit 50 or more HRs in a single season? (Dataframe, first name, last name, HRs, season)**

```python
#Aggregating
a24 = batting_data.groupby(['playerID','yearID','birthYear','nameFirst','nameLast'],as_index=False).sum().copy()
#Selecting players with HR greater than or equal to 50
hr = a24[a24['HR']>=50].copy()
#Calculating approximate age 
hr['Req_Age'] = hr['yearID'] - hr['birthYear']
#Displaying the youngest player to hit 50 or more HRs
display(hr[hr['Req_Age']== (hr['Req_Age'].min())][['playerID','nameFirst','nameLast','yearID','birthYear','HR','Req_Age']].reset_index(drop=True))
```

**25)Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe, first name, last name, season they eclipsed more than 300 HRs)**

```python
a25 = batting_data.groupby(['playerID', 'nameFirst', 'nameLast', 'birthYear', 'yearID']).sum().groupby(level=0).cumsum().reset_index().copy()
#Sorting
a25 = a25[a25['HR'] >= 300].sort_values(by = ['yearID', 'birthYear', 'playerID'])[['playerID', 'nameFirst', 'nameLast', 'birthYear', 'yearID', 'HR']]
#Finding the required age
a25['Req_Age'] = a25['yearID'] - a25['birthYear']
#Displaying the five youngest players to hit 300 or more HRs
display(a25.sort_values(by = 'Req_Age', ascending = True).head(5).reset_index(drop=True))
```
